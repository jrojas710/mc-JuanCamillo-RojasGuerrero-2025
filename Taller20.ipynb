```python
# Juan: importan librerias

import numpy as np  # Juan: se usa para operaciones numericas

import matplotlib.pyplot as plt  # Juan: se usa para graficar

# Juan: se definen los datos dados en el taller

x = np.array([0,1,2,3,4,5,6,7])  # Juan: valores de x

y = np.array([5.5,5.4,4.0,3.7,3.0,2.0,2.0,1.5])  # Juan: valores de y
```

```python
# Juan: se calcula el polinomio de segundo grado

coef = np.polyfit(x, y, 2)  # Juan: se obtienen los coeficientes a b y c

# Juan: se separan los coeficientes

a = coef[0]  # Juan: coeficiente cuadratico

b = coef[1]  # Juan: coeficiente lineal

c = coef[2]  # Juan: termino independiente

# Juan: se genera la funcion estimada

y_pred = a * x**2 + b * x + c  # Juan: valores del polinomio evaluado

print("Modelo resultante")

print(f"y = {a:.4f} x^2 + {b:.4f} x + {c:.4f}")
```

```python
# Juan: se calcula la suma de residuos

ss_res = np.sum((y - y_pred)**2)  # Juan: residuos al cuadrado

# Juan: se calcula la suma total

ss_tot = np.sum((y - np.mean(y))**2)  # Juan: diferencia con la media al cuadrado

# Juan: formula del coeficiente de determinacion

r2 = 1 - ss_res / ss_tot  # Juan: se obtiene R2

print(f"Coeficiente de determinacion R2 = {r2:.4f}")
```

```python
# Juan: se calcula la matriz de correlacion

mat = np.corrcoef(y, y_pred)  # Juan: matriz de correlacion

# Juan: se extrae solamente el coeficiente

corr = mat[0,1]  # Juan: valor del coeficiente r

print(f"Coeficiente de correlacion r = {corr:.4f}")
```

```python
# Juan: se crea un grafico con los puntos y el polinomio

plt.scatter(x, y, s=60, label='Datos')  # Juan: grafico de puntos

plt.plot(x, y_pred, label='Ajuste cuadratico')  # Juan: grafico del polinomio

# Juan: se agregan etiquetas

plt.xlabel('x')  # Juan: etiqueta eje x

plt.ylabel('y')  # Juan: etiqueta eje y

# Juan: se agrega un titulo

plt.title('Ajuste Polinomio de Segundo Grado')  # Juan: titulo del grafico

# Juan: se agrega una grilla

plt.grid(True)  # Juan: grilla para mejor visualizacion

# Juan: se agrega la leyenda

plt.legend()  # Juan: se muestran las etiquetas

# Juan: se muestra el grafico final
plt.show()
```
