```python
# Juan: se importan las librerias necesarias

import numpy as np  # Juan: se usa para calculos numericos

import matplotlib.pyplot as plt  # Juan: se usa para graficar

# Juan: se definen los datos dados en el enunciado

x = np.array([1, 2, 3, 4, 5, 6])  # Juan: valores de x

y = np.array([2.8, 5.1, 6.9, 9.2, 12.9, 18.1])  # Juan: valores de y

# Juan: se aplica transformacion logaritmica al modelo exponencial

# Juan: el modelo es y = a * e^(b*x)

# Juan: al aplicar logaritmo natural se obtiene ln(y) = ln(a) + b*x

ln_y = np.log(y)  # Juan: se calcula el logaritmo natural de y

# Juan: se calculan los parametros de la regresion lineal

n = len(x)  # Juan: se obtiene el numero de datos

sum_x = np.sum(x)  # Juan: suma de x

sum_ln_y = np.sum(ln_y)  # Juan: suma de ln(y)

sum_x_ln_y = np.sum(x * ln_y)  # Juan: suma de x * ln(y)

sum_x2 = np.sum(x ** 2)  # Juan: suma de x al cuadrado

# Juan: se aplica la formula de la pendiente b

b = (n * sum_x_ln_y - sum_x * sum_ln_y) / (n * sum_x2 - sum_x ** 2)  # Juan: pendiente del modelo

# Juan: se aplica la formula de la interseccion ln(a)

ln_a = (sum_ln_y - b * sum_x) / n  # Juan: interseccion del modelo

# Juan: se obtiene el valor de a aplicando la exponencial

a = np.exp(ln_a)  # Juan: constante a del modelo

# Juan: se imprime el modelo obtenido

print(f"Modelo exponencial: y = {a:.4f} * e^({b:.4f}x)")

# Juan: se generan valores predichos con el modelo

y_pred = a * np.exp(b * x)  # Juan: se calcula el valor estimado de y

# Juan: se grafican los resultados

plt.scatter(x, y, color='blue', label='Datos reales')  # Juan: puntos de los datos originales

plt.plot(x, y_pred, color='red', label='Modelo exponencial')  # Juan: linea del modelo

plt.title('Regresion exponencial por minimos cuadrados')  # Juan: titulo del grafico

plt.xlabel('x')  # Juan: etiqueta del eje x

plt.ylabel('y')  # Juan: etiqueta del eje y

plt.legend()  # Juan: se muestra la leyenda

plt.grid(True)  # Juan: se agrega la grilla

plt.show()  # Juan: se muestra el grafico

# Juan: se calcula el coeficiente de determinacion R^2

ss_res = np.sum((y - y_pred) ** 2)  # Juan: suma de residuos al cuadrado

ss_tot = np.sum((y - np.mean(y)) ** 2)  # Juan: suma total al cuadrado

r2 = 1 - (ss_res / ss_tot)  # Juan: formula de R cuadrado

# Juan: se imprime el valor de R^2

print(f"Coeficiente de determinacion R^2 = {r2:.4f}")
```
