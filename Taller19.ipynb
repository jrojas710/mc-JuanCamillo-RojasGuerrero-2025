```python
# Juan: se importan las librerias necesarias

import numpy as np  # Juan: se usa para los calculos numericos

import matplotlib.pyplot as plt  # Juan: se usa para graficar

# Juan: se definen los datos dados en el enunciado

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])  # Juan: valores de x

y = np.array([1.2, 3.3, 3.8, 4.0, 4.2, 4.4, 4.5, 4.6])  # Juan: valores de y
```

```python
# Juan: se aplica la regresion lineal

# Juan: el modelo es y = a * x + b

A = np.vstack([x, np.ones(len(x))]).T  # Juan: se crea la matriz A para la regresion lineal

a, b = np.linalg.lstsq(A, y, rcond=None)[0]  # Juan: se calcula la pendiente y la interseccion

# Juan: se imprime el modelo lineal

print(f"Modelo lineal: y = {a:.4f} * x + {b:.4f}")

# Juan: se generan los valores predichos con el modelo lineal

y_pred_lineal = a * x + b  # Juan: se calcula el valor estimado de y

# Juan: se grafica el modelo lineal

plt.scatter(x, y, color='blue', label='Datos reales')  # Juan: puntos de los datos originales

plt.plot(x, y_pred_lineal, color='red', label='Modelo lineal')  # Juan: linea del modelo lineal

plt.title('Regresion Lineal')  # Juan: titulo del grafico

plt.xlabel('x')  # Juan: etiqueta del eje x

plt.ylabel('y')  # Juan: etiqueta del eje y

plt.legend()  # Juan: se muestra la leyenda

plt.grid(True)  # Juan: se agrega la grilla

plt.show()  # Juan: se muestra el grafico

# Juan: se calcula el coeficiente de determinacion R^2 para el modelo lineal

ss_res = np.sum((y - y_pred_lineal) ** 2)  # Juan: suma de residuos al cuadrado

ss_tot = np.sum((y - np.mean(y)) ** 2)  # Juan: suma total al cuadrado

r2_lineal = 1 - (ss_res / ss_tot)  # Juan: formula de R cuadrado

# Juan: se imprime el valor de R^2

print(f"Coeficiente de determinacion R^2 (Lineal) = {r2_lineal:.4f}")
```

```python
# Juan: se aplica transformacion logaritmica al modelo exponencial

# Juan: el modelo es y = a * e^(b*x)

# Juan: al aplicar logaritmo natural se obtiene ln(y) = ln(a) + b*x

ln_y = np.log(y)  # Juan: se calcula el logaritmo natural de y

# Juan: se calculan los parametros de la regresion lineal para el modelo exponencial

A_exp = np.vstack([x, np.ones(len(x))]).T  # Juan: matriz A para la regresion exponencial

b_exp, ln_a_exp = np.linalg.lstsq(A_exp, ln_y, rcond=None)[0]  # Juan: calculo de los parametros

# Juan: se obtiene el valor de a aplicando la exponencial

a_exp = np.exp(ln_a_exp)  # Juan: constante a del modelo exponencial

# Juan: se imprime el modelo exponencial

print(f"Modelo exponencial: y = {a_exp:.4f} * e^({b_exp:.4f}x)")

# Juan: se generan los valores predichos con el modelo exponencial

y_pred_exp = a_exp * np.exp(b_exp * x)  # Juan: se calcula el valor estimado de y

# Juan: se grafica el modelo exponencial

plt.scatter(x, y, color='blue', label='Datos reales')  # Juan: puntos de los datos originales

plt.plot(x, y_pred_exp, color='green', label='Modelo exponencial')  # Juan: linea del modelo exponencial

plt.title('Regresion Exponencial')  # Juan: titulo del grafico

plt.xlabel('x')  # Juan: etiqueta del eje x

plt.ylabel('y')  # Juan: etiqueta del eje y

plt.legend()  # Juan: se muestra la leyenda

plt.grid(True)  # Juan: se agrega la grilla

plt.show()  # Juan: se muestra el grafico

# Juan: se calcula el coeficiente de determinacion R^2 para el modelo exponencial

ss_res_exp = np.sum((y - y_pred_exp) ** 2)  # Juan: suma de residuos al cuadrado

r2_exp = 1 - (ss_res_exp / ss_tot)  # Juan: formula de R cuadrado

# Juan: se imprime el valor de R^2

print(f"Coeficiente de determinacion R^2 (Exponencial) = {r2_exp:.4f}")
```

```python
# Juan: se aplica la regresion por potencias

# Juan: el modelo es y = a * x^b

ln_x = np.log(x)  # Juan: se calcula el logaritmo natural de x

ln_y = np.log(y)  # Juan: se calcula el logaritmo natural de y

# Juan: se calculan los parametros de la regresion lineal

A_pow = np.vstack([ln_x, np.ones(len(x))]).T  # Juan: matriz A para la regresion por potencias

b_pow, ln_a_pow = np.linalg.lstsq(A_pow, ln_y, rcond=None)[0]  # Juan: calculo de los parametros

# Juan: se obtiene el valor de a aplicando la exponencial

a_pow = np.exp(ln_a_pow)  # Juan: constante a del modelo por potencias

# Juan: se imprime el modelo por potencias

print(f"Modelo por potencias: y = {a_pow:.4f} * x^{b_pow:.4f}")

# Juan: se generan los valores predichos con el modelo por potencias

y_pred_pow = a_pow * x ** b_pow  # Juan: se calcula el valor estimado de y

# Juan: se grafica el modelo por potencias

plt.scatter(x, y, color='blue', label='Datos reales')  # Juan: puntos de los datos originales

plt.plot(x, y_pred_pow, color='orange', label='Modelo por potencias')  # Juan: linea del modelo por potencias

plt.title('Regresion por Potencias')  # Juan: titulo del grafico

plt.xlabel('x')  # Juan: etiqueta del eje x

plt.ylabel('y')  # Juan: etiqueta del eje y

plt.legend()  # Juan: se muestra la leyenda

plt.grid(True)  # Juan: se agrega la grilla

plt.show()  # Juan: se muestra el grafico

# Juan: se calcula el coeficiente de determinacion R^2 para el modelo por potencias

ss_res_pow = np.sum((y - y_pred_pow) ** 2)  # Juan: suma de residuos al cuadrado

r2_pow = 1 - (ss_res_pow / ss_tot)  # Juan: formula de R cuadrado

# Juan: se imprime el valor de R^2

print(f"Coeficiente de determinacion R^2 (Potencias) = {r2_pow:.4f}")
```

```python
# Juan: se aplica la regresion por razon de crecimiento

# Juan: el modelo es y = a * b^x

ln_y = np.log(y)  # Juan: se calcula el logaritmo natural de y

# Juan: se calculan los parametros de la regresion lineal

A_growth = np.vstack([x, np.ones(len(x))]).T  # Juan: matriz A para la regresion por razon de crecimiento

b_growth, ln_a_growth = np.linalg.lstsq(A_growth, ln_y, rcond=None)[0]  # Juan: calculo de los parametros

# Juan: se obtiene el valor de a aplicando la exponencial

a_growth = np.exp(ln_a_growth)  # Juan: constante a del modelo por razon de crecimiento

# Juan: se imprime el modelo por razon de crecimiento

print(f"Modelo por razon de crecimiento: y = {a_growth:.4f} * b^{b_growth:.4f}")

# Juan: se generan los valores predichos con el modelo por razon de crecimiento

y_pred_growth = a_growth * np.exp(b_growth * x)  # Juan: se calcula el valor estimado de y

# Juan: se grafica el modelo por razon de crecimiento

plt.scatter(x, y, color='blue', label='Datos reales')  # Juan: puntos de los datos originales

plt.plot(x, y_pred_growth, color='purple', label='Modelo por razon de crecimiento')  # Juan: linea del modelo por razon de crecimiento

plt.title('Regresion por Razon de Crecimiento')  # Juan: titulo del grafico

plt.xlabel('x')  # Juan: etiqueta del eje x

plt.ylabel('y')  # Juan: etiqueta del eje y

plt.legend()  # Juan: se muestra la leyenda

plt.grid(True)  # Juan: se agrega la grilla

plt.show()  # Juan: se muestra el grafico

# Juan: se calcula el coeficiente de determinacion R^2 para el modelo por razon de crecimiento

ss_res_growth = np.sum((y - y_pred_growth) ** 2)  # Juan: suma de residuos al cuadrado

r2_growth = 1 - (ss_res_growth / ss_tot)  # Juan: formula de R cuadrado

# Juan: se imprime el valor de R^2

print(f"Coeficiente de determinacion R^2 (Razon de Crecimiento) = {r2_growth:.4f}")
```
