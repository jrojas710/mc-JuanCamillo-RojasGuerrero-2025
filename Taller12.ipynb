```python
# Juan: Se importan librerias

import math

import pandas as pd
```

```python
# Juan: Se definen los valores de la base y el punto de evaluacion

a = 0.8  # Juan: Valor base donde se desarrolla la serie

x = 0.805  # Juan: Punto donde se quiere aproximar la funcion

orden_max = 15  # Juan: Maximo orden de la serie de Taylor

# Juan: Valor real de la funcion para calcular errores

valor_real = math.exp(-x)

valor_real
```

```python
# Juan: Se define funcion para calcular la aproximación de Taylor

def taylor_expx_neg(x, a, n):

    suma = 0

    for k in range(n+1):

        derivada_en_a = ((-1)**k) * math.exp(-a)  # Juan: k-ésima derivada evaluada en a

        termino = (derivada_en_a / math.factorial(k)) * ((x - a)**k)  # Juan: Termino de la serie

        suma += termino  # Juan: Se acumula la suma

    return suma
```

```python
# Juan: Se crea listas para guardar los resultados

ordenes = []

aprox = []

errores = []

for n in range(orden_max+1):

    valor_aprox = taylor_expx_neg(x, a, n)  # Juan: Se calcula la aproximacion

    error = abs((valor_real - valor_aprox) / valor_real) * 100  # Juan: Error relativo porcentual

    ordenes.append(n)

    aprox.append(valor_aprox)

    errores.append(error)

# Juan: Se crea un DataFrame para mostrar resultados

df = pd.DataFrame({

    'Orden': ordenes,

    'Aproximación': aprox,

    'Error %': errores

})

df
```

```python
# Juan: Graficacion

import matplotlib.pyplot as plt

plt.plot(ordenes, errores, marker='o')

plt.xlabel('Orden de Taylor')  # Juan: Eje X es el orden

plt.ylabel('Error relativo %')  # Juan: Eje Y es el error

plt.title('Error relativo porcentual de la aproximación')

plt.grid(True)

plt.show()
```
