```python
# Juan: Importan librerias

import numpy as np

# Juan: se importa matplotlib para graficar

import matplotlib.pyplot as plt

# Juan: se definen los datos de x y y segun la tabla

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])

y = np.array([12, 8, 10, 4, 6, 3, 2, -1])

# Juan: se calculan los promedios de x y y

x_mean = np.mean(x)

y_mean = np.mean(y)

# Juan: se aplican las formulas de regresion lineal por minimos cuadrados

num = np.sum((x - x_mean) * (y - y_mean))

den = np.sum((x - x_mean) ** 2)

m = num / den

# Juan: se calcula la intersección con el eje y

b = y_mean - m * x_mean

# Juan: se muestra la ecuacion de la recta

print(f"Ecuacion de la recta: y = {m:.2f}x + {b:.2f}")

# Juan: se calculan los valores predichos de y

y_pred = m * x + b

# Juan: se crea la grafica

plt.scatter(x, y, color='blue', label='Datos originales')

# Juan: se grafica la linea de regresión

plt.plot(x, y_pred, color='red', label='Línea ajustada')

# Juan: se añaden los titulos y etiquetas

plt.title('Regresiion Lineal por Minimos Cuadrados')

plt.xlabel('x')

plt.ylabel('y')

plt.legend()

# Juan: se muestra la grafica

plt.show()
```
