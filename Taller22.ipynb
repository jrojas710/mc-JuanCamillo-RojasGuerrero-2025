```python
# Juan: se define clase Nodo

class Nodo:

    def __init__(self, valor):  # Juan: constructor

        self.valor = valor  # Juan: guarda valor

        self.izquierdo = None  # Juan: hijo izquierdo

        self.derecho = None  # Juan: hijo derecho
```

```python
# Juan: se define clase ArbolBinario

class ArbolBinario:

    def __init__(self):  # Juan: constructor

        self.raiz = None  # Juan: raiz del arbol


    def agregarValor(self, valor):  # Juan: metodo p√∫blico agregar

        if self.raiz is None:  # Juan: si arbol vacio

            self.raiz = Nodo(valor)  # Juan: crea raiz
        else:
            self._insertar(self.raiz, valor)  # Juan: inserta recursivo


    def _insertar(self, nodo, valor):  # Juan: metodo interno insertar

        if valor < nodo.valor:  # Juan: va por izquierda

            if nodo.izquierdo is None:

                nodo.izquierdo = Nodo(valor)  # Juan: asigna nuevo

            else:

                self._insertar(nodo.izquierdo, valor)  # Juan: recurre izquierda

        else:  # Juan: va por derecha

            if nodo.derecho is None:

                nodo.derecho = Nodo(valor)  # Juan: asigna nuevo

            else:

                self._insertar(nodo.derecho, valor)  # Juan: recurre derecha


    def buscarValor(self, valor):  # Juan: metodo buscar

        return self._buscar(self.raiz, valor)  # Juan: usa recursivo


    def _buscar(self, nodo, valor):  # Juan: metodo interno buscar

        if nodo is None:  # Juan: caso base

            return False

        if nodo.valor == valor:  # Juan: encontrado

            return True

        if valor < nodo.valor:  # Juan: va izquierda

            return self._buscar(nodo.izquierdo, valor)

        else:  # Juan: va derecha

            return self._buscar(nodo.derecho, valor)


    def imprimirValores(self):  # Juan: imprime en orden

        self._inorden(self.raiz)


    def _inorden(self, nodo):  # Juan: recorrido inorden

        if nodo is not None:

            self._inorden(nodo.izquierdo)

            print(nodo.valor)

            self._inorden(nodo.derecho)
```

```python
# Juan: se crea el arbol

arbol = ArbolBinario()


# Juan: se generan 20 numeros aleatorios sin repetir

numeros = random.sample(range(1, 101), 20)

print("Numeros generados:", numeros)



# Juan: se insertan en el arbol

for n in numeros:

    arbol.agregarValor(n)



# Juan: menu

while True:

    print("\n1. Buscar valor")

    print("2. Imprimir en orden")

    print("3. Salir")

    op = input("Opcion: ")


    if op == "1":

        v = int(input("Valor a buscar: "))

        if arbol.buscarValor(v):

            print("El valor esta en el arbol")


        else:

            print("El valor no esta en el arbol")


    elif op == "2":

        print("Valores del arbol:")

        arbol.imprimirValores()



    elif op == "3":


        break


    else:

        print("Opcion invalida")
```
