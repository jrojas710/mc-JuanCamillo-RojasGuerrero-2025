**Puntos:**
| x | f(x) |
|---|------|
| 0 | 1 |
| 1 | -0.5 |
| 2 | -1 |
| 3 | 2.5 |
| 4 | 0.5 |

```python
# Juan: Importamos las librerias necesarias

import numpy as np  # Juan: Numpy para calculos numericos

import matplotlib.pyplot as plt  # Juan: Matplotlib para graficar
```

```python
# Juan: Definimos los puntos dados

x = np.array([0, 1, 2, 3, 4])  # Juan: Coordenadas en x

y = np.array([1, -0.5, -1, 2.5, 0.5])  # Juan: Valores de la función f(x)
```

```python
# Juan: Funcion para calcular el polinomio de Lagrange

def lagrange_interpolacion(x, y, x_eval):  # Juan: x_eval es el valor donde queremos evaluar el polinomio

    total = 0  # Juan: Inicializamos la suma total en cero

    n = len(x)  # Juan: Obtenemos el número de puntos

    for i in range(n):  # Juan: Recorremos cada punto

        term = y[i]  # Juan: Empezamos con el valor f(x_i)

        for j in range(n):  # Juan: Segundo bucle para construir los términos del producto

            if i != j:  # Juan: Evitamos dividir por cero

                term *= (x_eval - x[j]) / (x[i] - x[j])  # Juan: Formula del termino de Lagrange

        total += term  # Juan: Sumamos cada término al total

    return total  # Juan: Retornamos el valor del polinomio en x_eval
```

```python
# Juan: Probamos la función en varios puntos del rango

x_eval = np.linspace(min(x), max(x), 100)  # Juan: Creamos 100 puntos entre el mínimo y máximo de x

y_eval = [lagrange_interpolacion(x, y, xi) for xi in x_eval]  # Juan: Evaluamos el polinomio en esos puntos
```

```python
# Juan: Graficamos los puntos y el polinomio interpolante

plt.figure(figsize=(8,6))  # Juan: Definimos el tamaño de la figura

plt.plot(x_eval, y_eval, label='Polinomio de Lagrange', color='blue')  # Juan: Linea del polinomio

plt.scatter(x, y, color='red', label='Puntos dados')  # Juan: Puntos originales

plt.title('Interpolación de Lagrange')  # Juan: Titulo del gráfico

plt.xlabel('x')  # Juan: Etiqueta eje x

plt.ylabel('f(x)')  # Juan: Etiqueta eje y

plt.legend()  # Juan: Mostramos la leyenda

plt.grid(True)  # Juan: Activamos la grilla

plt.show()  # Juan: Mostramos la gráfica
```

```python
# Juan: Evaluamos el polinomio en un punto específico por ejemplo x = 2.5

valor = lagrange_interpolacion(x, y, 2.5)  # Juan: Evaluamos en 2.5

print('El valor interpolado en x = 2.5 es:', valor)  # Juan: Mostramos el resultado
```
