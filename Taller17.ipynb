```python
# se importa numpy pandas y matplotlib

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

## datos

```python
# se crea el arreglo de datos

x = np.array([0, 2, 4, 6, 8, 10, 12], dtype=float)

y = np.array([0, 1, 4, 7, 9, 18, 27], dtype=float)

# se arma el dataframe

df = pd.DataFrame({'x': x, 'y': y})

df
```

## ajuste lineal y metricas


```python
# se calcula estadisticos basicos

n = len(x)

x_bar = x.mean()

y_bar = y.mean()

# se calcula sumas de cuadrados

Sxx = np.sum((x - x_bar) ** 2)

Sxy = np.sum((x - x_bar) * (y - y_bar))

# se calcula pendiente e intercepcion

b1 = Sxy / Sxx

b0 = y_bar - b1 * x_bar

# se calcula valores ajustados y residuos

y_hat = b0 + b1 * x

resid = y - y_hat

# se calcula sse y sst

SSE = np.sum(resid ** 2)

SST = np.sum((y - y_bar) ** 2)

# se calcula sy desviacion estandar muestral de y

sy = np.sqrt(SST / (n - 1))

# se calcula sy barra x error estandar de la estimacion

syx = np.sqrt(SSE / (n - 2))

# se calcula r y r2

r = Sxy / np.sqrt(Sxx * SST)

r2 = r ** 2
```

```python
# se imprimen resultados

print(f"b0 intercepcion  {b0:.6f}")

print(f"b1 pendiente  {b1:.6f}")

print(f"sy desv estandar de y  {sy:.6f}")

print(f"sy_x error estandar de la estimacion  {syx:.6f}")

print(f"r coef de correlacion  {r:.6f}")

print(f"r2 coef de determinacion  {r2:.6f}")
```

## grafica de dispersion y recta de ajuste

```python
# se prepara la grafica

plt.figure()

plt.scatter(x, y, label='datos')

x_line = np.linspace(x.min(), x.max(), 100)

y_line = b0 + b1 * x_line

plt.plot(x_line, y_line, label='ajuste lineal')

plt.xlabel('x')

plt.ylabel('y')

plt.title('grafica de dispersion y ajuste lineal')

plt.legend()

plt.grid(True)

plt.show()
```

## verificacion

```python
# se verifica con numpy

b1_np, b0_np = np.polyfit(x, y, 1)

r_np = np.corrcoef(x, y)[0, 1]

print(f"b0 np  {b0_np:.6f}  b1 np  {b1_np:.6f}")

print(f"r np  {r_np:.6f}")
```
