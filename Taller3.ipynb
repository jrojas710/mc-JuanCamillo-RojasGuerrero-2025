```python
# Juan: Lista de los únicos dos valores logicos posibles en algebra booleana: True y False

valores_logicos = [True, False]

# Juan: Funcion para convertir un valor booleano a su representación textual

def boolean(valor):

    # Juan: Si el valor es verdadero, se retorna "True", si no, "False"

    return "True" if valor else "False"

# Juan: Esta funcion imprime una tabla comparando dos expresiones lógicas usando 2 o 3 variables

def imprimir_tabla(expr1, expr2=None, variables=3, titulo=""):

    # Juan: Imprime el titulo de la tabla actual (nombre de la ley que se va a verificar)

    print(f"\n --- {titulo} ---")

    # Juan: Si la tabla es de 2 variables (como x, y)

    if variables == 2:
        # Juan: Se itera por todas las combinaciones posibles de x y y

        for x in valores_logicos:

            for y in valores_logicos:

                # Juan: Evaluamos ambas expresiones lógicas con los valores actuales

                resultado_1 = expr1(x, y)

                resultado_2 = expr2(x, y) if expr2 else None

                # Juan: Imprimimos los resultados en formato tabla, dependiendo si hay una o dos expresiones

                if expr2:

                    print(f"{boolean(x):^7}|{boolean(y):^7}|{boolean(resultado_1):^7}|{boolean(resultado_2):^7}|")

                else:

                    print(f"{boolean(x):^7}|{boolean(y):^7}|{boolean(resultado_1):^7}|")

    else:
        # Juan: Si la tabla es de 3 variables (como x, y, z)

        for x in valores_logicos:

            for y in valores_logicos:

                for z in valores_logicos:

                    # Juan: Evaluamos ambas expresiones con los valores actuales

                    resultado_1 = expr1(x, y, z)

                    resultado_2 = expr2(x, y, z) if expr2 else None

                    # Juan: Imprimimos los resultados de las operaciones en la tabla

                    if expr2:

                        print(f"{boolean(x):^7}|{boolean(y):^7}|{boolean(z):^7}|{boolean(resultado_1):^7}|{boolean(resultado_2):^7}|")

                    else:

                        print(f"{boolean(x):^7}|{boolean(y):^7}|{boolean(z):^7}|{boolean(resultado_1):^7}|")

# Juan: Función que demuestra la ley asociativa

def ley_asociativa():

    # Juan: Primera ley asociativa: (x OR y) OR z = x OR (y OR z)

    imprimir_tabla(

        lambda x, y, z: (x or y) or z,

        lambda x, y, z: x or (y or z),

        3,

        "Primera ley asociativa"
    )

    # Juan: Segunda ley asociativa: (x AND y) AND z = x AND (y AND z)

    imprimir_tabla(

        lambda x, y, z: (x and y) and z,

        lambda x, y, z: x and (y and z),

        3,

        "Segunda ley asociativa"
    )

# Juan: Función que demuestra la ley conmutativa

def ley_conmutativa():

    # Juan: Primera ley conmutativa: x OR y = y OR x

    imprimir_tabla(

        lambda x, y: x or y,

        lambda x, y: y or x,

        2,

        "Primera ley conmutativa"

    )

    # Juan: Segunda ley conmutativa: x AND y = y AND x

    imprimir_tabla(

        lambda x, y: x and y,

        lambda x, y: y and x,

        2,

        "Segunda ley conmutativa"

    )

# Juan: Función que demuestra la ley distributiva

def ley_distributiva():

    # Juan: Primera ley distributiva: x AND (y OR z) = (x AND y) OR (x AND z)

    imprimir_tabla(

        lambda x, y, z: x and (y or z),

        lambda x, y, z: (x and y) or (x and z),

        3,

        "Primera ley distributiva"

    )

    # Juan: Segunda ley distributiva: x OR (y AND z) = (x OR y) AND (x OR z)

    imprimir_tabla(

        lambda x, y, z: x or (y and z),

        lambda x, y, z: (x or y) and (x or z),

        3,

        "Segunda ley distributiva"

    )

# Juan: Función que demuestra la ley de identidad

def ley_identidad():

    # Juan: Primera ley identidad: x OR False = x

    imprimir_tabla(

        lambda x, y: x or False,

        None,

        2,

        "Primera ley identidad"

    )

    # Juan: Segunda ley identidad: x AND True = x

    imprimir_tabla(

        lambda x, y: x and True,

        None,

        2,

        "Segunda ley identidad"

    )

# Juan: Función que demuestra la ley de complementos

def ley_complementos():

    # Juan: Primera ley de complementos: x OR NOT x = True

    imprimir_tabla(

        lambda x, y: x or not x,

        None,

        2,

        "Primera ley complementos"

    )

    # Juan: Segunda ley de complementos: x AND NOT x = False

    imprimir_tabla(

        lambda x, y: x and not x,

        None,

        2,

        "Segunda ley complementos"

    )

# Juan: Función principal del programa, que ejecuta todas las demostraciones

def main():

    # Juan: Se llama una por una a las funciones que imprimen cada ley

    ley_asociativa()

    ley_conmutativa()

    ley_distributiva()

    ley_identidad()

    ley_complementos()

if __name__ == "__main__":

    main()
```
